```python
from IPython.display import display
from google.colab import files
import pandas as pd
import io
import matplotlib.pyplot as plt

# Upload file
uploaded = files.upload()

# Load Excel file
file_name = list(uploaded.keys())[0]
df = pd.read_excel(io.BytesIO(uploaded[file_name]))

# Display first few rows
display(df.head())
```

```python
# Plot: Number of graduates by major
graduates_count = df['Major'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12, 9))
graduates_count.plot(kind='bar', color='cornflowerblue')
plt.title('Number of Graduates by Major', fontsize=30, fontweight='bold')
plt.xlabel('Major', fontsize=25)
plt.ylabel('Number of Graduates', fontsize=25)

plt.xticks(rotation=45, fontsize=20)
plt.yticks(fontsize=20)

plt.tight_layout()
plt.show()

# Show major with the most graduates
most_grads_major = graduates_count.idxmax()
most_grads_count = graduates_count.max()
from IPython.display import display, HTML
# Bigger font size using inline CSS
display(HTML(f'<div style="font-size:28px; color: darkblue; font-weight: bold;">Major with the most graduates: {most_grads_major} ({most_grads_count})</div>'))
```

```python
# Plot: Average monthly salary by major
avg_salary_by_major = df.groupby('Major')['Monthy Salary'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 9))
avg_salary_by_major.plot(kind='bar', color='mediumseagreen')  # Changed bar color to green
plt.title('Average Monthly Starting Salary by Major', fontsize=30, fontweight='bold')
plt.xlabel('Major', fontsize=25)
plt.ylabel('Average Monthly Salary ($)', fontsize=25)

plt.xticks(rotation=45, fontsize=20)
plt.yticks(fontsize=20)

plt.tight_layout()
plt.show()

# Show major with the highest average salary
highest_salary_major = avg_salary_by_major.idxmax()
highest_avg_salary = avg_salary_by_major.max()

from IPython.display import display, HTML

# Display result in large green font
display(HTML(f'<div style="font-size:28px; color: darkgreen; font-weight: bold;">Major with the highest average monthly salary: {highest_salary_major} (${highest_avg_salary:.2f})</div>'))
```
